```python
# 임대보증금, 임대료 결측치 채우기
train[['임대보증금', '임대료']] = train[['임대보증금', '임대료']].fillna(0)
test[['임대보증금', '임대료']] = test[['임대보증금', '임대료']].fillna(0)
```

```python
# 지하철역, 보스정류장 결측치 채우기 
cols = ['지하철', '버스']
train[cols] = train[cols].fillna(0)
test[cols] = test[cols].fillna(0)
```
